### 環境準備（直接執行即可）
會載入 Day 2 用過的中文 BERT 模型，因為剛剛下載過，這次會很快。

```python
import subprocess, sys
subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'transformers', '-q'])

from transformers import pipeline
print('正在載入兩個中文 BERT 模型...')
# 1) 情緒分析（基於京東商城評論微調）
classifier = pipeline("sentiment-analysis", model="uer/roberta-base-finetuned-jd-binary-chinese")
# 2) 克漏字接話
unmasker = pipeline("fill-mask", model="bert-base-chinese")
print('準備完成！可以開始玩了。')
```

### 第一站：你的中文情緒偵測器
先看一個範例，再換成**你自己的句子**。

```python
def detect_emotion(text):
    res = classifier(text)[0]
    label_str = res['label'].lower()
    is_positive = 'positive' in label_str or label_str in ('label 1', 'label_1')
    mood = "正面 (開心)" if is_positive else "負面 (不開心)"
    print(f"句子：『{text}』")
    print(f"  → AI 判斷：{mood}（信心 {res['score']*100:.1f}%）\n")

# 範例
detect_emotion("這兩天的 AI 營隊真是太精彩了！")

# TODO: 把下面兩句換成「你自己」最近的心情或想說的話，再執行這一格
# 例如：detect_emotion("今天終於訓練出我的第一個模型，超有成就感")
# ??? (請在此填寫你的答案)
```

### 第二站：和 BERT 玩接話（克漏字）
`[MASK]` 代表被遮住的字，看看 BERT 猜得準不準。

```python
def guess_blank(sentence):
    print(f"題目：{sentence}")
    for i, res in enumerate(unmasker(sentence)[:3], 1):
        print(f"  猜測 {i}：{res['token_str']}（信心 {res['score']*100:.1f}%）")
    print()

guess_blank("我最喜歡的科目是[MASK]。")
guess_blank("學會 AI 之後，我想用它來解決[MASK]問題。")
```
